```python
def act(h,L):
    result = Set(h(L[i]) for i in range(0,L.cardinality()))
    return result

n = 6  #number of vertices (Remember to also change numVertices in the "nauty_geng" argument)
k = 3  #Size of subsets

Graphs = graphs.nauty_geng("6 5:15 -c") #Generates all connected graphs on specified numVertices and numEdges ***Must input actual value of vertices and edges in argument: ("numVertices numEdges:numEdges -c").  "-c" restricts to connected graphs.
kSubsets = Subsets(range(0,n), k)  #All k-subsets of {0,1,...,n-1}
for G in Graphs:
    G.show(vertex_labels=True, edge_labels=False, graph_border=True, figsize=[4,4])  #prints out the graph
    print("Set", "|D|", "|I|", "|Sym(G)|")
    SymmetryGroup = G.automorphism_group()
    disjointActionDensity = 0 #initializes the density statistic
    for K in kSubsets:
        D = [] #disjoint producing actions
        I = [] #intersecting producing actions
        #print(K)
        for g in SymmetryGroup:
            if act(g,K).intersection(K).is_empty() == false: #if gK intersects K, then put it in I.  Otw put it in D.
                I.append(g)
            else:
                D.append(g)
        print(K,len(D),len(I),len(SymmetryGroup)) #Useful statistics for each k-subset
        disjointActionDensity += len(D)/len(SymmetryGroup) #summing up D densities for each k-subset to get average below
    disjointActionDensity = disjointActionDensity/len(kSubsets) #calculating D density average over all k-subsets
    print("Average Disjoint Action Density:", disjointActionDensity)
            
    
        
```
